```python
# Data Visialization
# Categorical plot
import matplotlib.pyplot as plt

import seaborn as sns

sns.set_theme(style="ticks",color_codes=True)

titanic = sns.load_dataset("titanic")
p1 =sns.catplot(x="sex",y="survived",data=titanic,hue='class',kind='bar')
plt.show()
# p1 =sns.countplot(x="sex",data=titanic,hue='class')
# p1.set_title('plot for couting')
plt.show()
```

```python
# Data Visialization
# Count Plot``
import matplotlib.pyplot as plt

import seaborn as sns

sns.set_theme(style="ticks",color_codes=True)

titanic = sns.load_dataset("titanic")
p1 =sns.countplot(x="sex",data=titanic,hue='class')
p1.set_title('plot for couting')
plt.show()
```

```python
# Data Visialization
# Scatter Plot``
import matplotlib.pyplot as plt

import seaborn as sns

sns.set_theme(style="ticks",color_codes=True)

titanic = sns.load_dataset("titanic")
p1 =sns.FacetGrid(titanic,hue='alone',row='sex')

p1 = (p1.map(plt.scatter,'age','fare').add_legend())
plt.show()
```
