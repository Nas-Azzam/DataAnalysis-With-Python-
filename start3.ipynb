```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

d = sns.load_dataset('tips')
sns.boxplot(x='day', y='tip', data=d,showmeans=True,meanprops={'marker':'*','markersize':'12'})
plt.xlabel('Day in week'),
plt.ylabel('Tip given per day'),
plt.title('TIPS VS DAYS',size = 20,weight='bold')
plt.show() 
d.head()
```
